```python
import pandas as pd
df = pd.read_csv("salaries_by_college_major.csv")
clean_df = df.dropna()
highest_value = clean_df.sort_values("Mid-Career 90th Percentile Salary", ascending=False)
highest_value[["Undergraduate Major", "Mid-Career 90th Percentile Salary"]].head()
```

```python
spread_col = clean_df["Mid-Career 90th Percentile Salary"] - clean_df["Mid-Career 10th Percentile Salary"]
# clean_df.insert(1, "Spread", spread_col )
difference = clean_df.sort_values("Spread", ascending=False)
difference[["Undergraduate Major", "Spread"]].head()
```

```python
groups = clean_df.groupby("Group").count()
print(groups)
```

```python
mean = clean_df.groupby("Group").mean()
print(mean)
```
